```python
in_file = "data.csv"
out_file = "tem10y.csv"

# CSV 파일을 한 줄 씩 읽어 들이기 ---(*1)
with open(in_file, "rt", encoding="EUC_KR") as fr:
    lines = fr.readlines()

# 기존의 데이터를 분리해서 가공하기 ---(*2)
lines = ["연,월,일,기온,품질,균질\n"] + lines[5:]
lines = map(lambda v: v.replace('/', ','), lines)
result = "".join(lines).strip()
print(result)

# 결과를 파일에 출력하기 ---(*3)
with open(out_file, "wt", encoding="utf-8") as fw:
    fw.write(result)
    print("saved.")
```

```python
import pandas as pd

#Pandas로 파일 읽어 들이기
df = pd.read_csv("tem10y.csv", encoding = "utf-8")

#날짜별로 기온을 리스트에 넣기
md = {}

for i, row in df.iterrows():
    m, d, v = (int(row['월']), int(row['일']), float(row['기온']))
    key =str(m)+"/"+str(d)
    
    if not(key in md):
        md[key] = []
    
    md[key] = md[key] + [v]

#날짜별 평균 구하기
avs = {}

for key in md:
    v = avs[key] = sum(md[key])/len(md[key])
    print("{0} : {1}".format(key,v))
```

```python
#특정 날짜의 평균기온 출력
print(avs["11/3"])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#CSV 파일 읽어 들이기
df= pd.read_csv("tem10y.csv",encoding = "utf-8")

#월별 평균구하기
g = df.groupby(['월'])['기온']
gg = g.sum() /g.count()

#결과 출력하기
print(gg)
gg.plot()
plt.savefig("tem-month-avg.png")
plt.show()
```

```python
#기온이 30도 넘는 날 구하기, Pandas filter

hot_bool = (df["기온"] > 30)

#데이터 추출하기
hot = df[hot_bool]

#연별로 세기
cnt = hot.groupby(["연"])["월"].count()

#출력하기
print(cnt)
cnt.plot()
plt.savefig("tem-over30.png")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#기온 데이터 읽어 들이기
df = pd.read_csv('tem10y.csv', encoding="utf-8")

#데이터를 학습전용과 테스트전용으로 분리하기
train_year = (df["연"] <= 2015)
test_year = (df["연"] >= 2016)
interval = 6

#과거 6일의 데이터를 기반으로 학습할 데이터 만들기
def make_data(data):
    x = [] #학습 데이터
    y = [] #결과
    temps= list(data["기온"])
    for i in range(len(temps)):
        if (i < interval):
            continue
        
        y.append(temps[i])
        xa = []
        
        for p in range(interval):
            d = i + p - interval
            xa.append(temps[d])
        x.append(xa)
        
    return (x, y)
    
train_x, train_y = make_data(df[train_year])
test_x, test_y = make_data(df[test_year])

#직선 회귀 분석하기
lr = LinearRegression(normalize=True)
lr.fit(train_x, train_y) # 학습하기
pre_y = lr.predict(test_x) # 예측하기

#결과를 그래프로 그리기
plt.figure(figsize= (10, 6), dpi = 100)
plt.plot(test_y, c ='r')
plt.plot(pre_y, c = 'b')
plt.savefig('weather-tem-lr.png')
plt.show()
```
